```python
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
gp = GaussianProcessRegressor(kernel=RBF(length_scale=5))
```

```python
n = 2
t = 50
q = 15
r = 2
```

```python
x = np.linspace(1, t, t)
curves = gp.sample_y(x.reshape((t, 1)), n * r)
```

```python
F = np.zeros((n * t, r))
for i in range(n):
    for j in range(r):
        F[(i * t):(i * t + t), j] = curves[:, (i * r + j)]
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
for j in range(r):
    plt.plot(x, F[:t, j])
plt.title('True Signal')
plt.xlabel('Time')
plt.show()
```

```python
loading = np.random.normal(0, 1, [r, q])
Y_true = np.matmul(F, loading)
Y = Y_true + np.random.normal(0, 0.5, [n * t, q])
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
for j in range(q):
    plt.plot(x, Y[:t, j])
plt.title('Observed Data')
plt.xlabel('Time')
plt.show()
```

```python
from factorgp import FactorGP
from inference import run_gibbs
```

```python
dims = [n, t, q, r]  # n, t, q, r
model = FactorGP(dims)
model.update_conditional_latent(Y)
```

```python
results = run_gibbs(Y, model, 100, 50, 0.5, verbose=True)
```

```python
F_sample, loading_sample, variance_sample, theta_sample, traces_hist, mse_history = results
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
plt.plot(mse_history)
plt.title('Model MSE')
plt.xlabel('Iteration')
plt.show()
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
for i in range(100):
    plt.plot(x, F_sample[i, :50, 0], color='orange', alpha=0.1)
    plt.plot(x, F_sample[i, :50, 1], color='blue', alpha=0.1)
plt.title('Latent Posterior')
plt.xlabel('Time')
plt.show()
```

```python
fig = plt.figure(figsize=(4, 3), dpi=150)
plt.hist(theta_sample[:, 0], bins=10, alpha=0.5, color='orange')
plt.hist(theta_sample[:, 1], bins=10, alpha=0.5, color='blue')
plt.title('Length Scale Posterior')
plt.xlabel('Factor')
plt.show()
```
